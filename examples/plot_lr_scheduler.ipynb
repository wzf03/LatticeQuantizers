```python
import matplotlib.pyplot as plt

from lattice_quantizer.lr_scheduler import CosineLR, FactorizedLR, RatioLR
```

```python
def plot_lr_scheduler(scheduler, num_steps, **kwargs):
    sch = scheduler(**kwargs).get_lr_scheduler(num_steps)

    lrs = [sch(step) for step in range(num_steps)]

    plt.plot(lrs)
```

```python
plot_lr_scheduler(
    CosineLR, 10000, max_update=5000, warmup_steps=1000, base_lr=0.01, final_lr=0.0
)
```

```python
plot_lr_scheduler(
    FactorizedLR, 10000, warmup_steps=1000, peak_lr=0.1, end_lr=0.001, factor=0.999
)
```

```python
plot_lr_scheduler(RatioLR, 1_000_000, ratio=200, peak_lr=0.005)
```
